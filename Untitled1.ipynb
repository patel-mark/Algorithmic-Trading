```python
# --- CELL 2: IMPORTS ---
import pandas as pd
import numpy as np
import backtrader as bt
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import pandas_ta as ta
%matplotlib inline
```

```python
# --- CELL 3: DATA LOADING ---
def load_data():
    df = pd.read_csv('trading_dataset.csv', index_col=0, parse_dates=True)
    
    # Validate indicator calculations
    df['SPY_RSI_14'] = ta.rsi(df['SPY_Close'], length=14)
    df['SPY_MACD_12_26'] = ta.macd(df['SPY_Close'], fast=12, slow=26)['MACD_12_26_9']
    df['QQQ_RSI_14'] = ta.rsi(df['QQQ_Close'], length=14)
    
    print("Data Summary (Last 5 Rows):")
    display(df[['SPY_Close', 'VIXCLS', 'SPY_RSI_14', 'SPY_MACD_12_26']].tail())
    return df

full_data = load_data()
```

```python
# --- CELL 4: BACKTRADER DATA FEED ---
class IndicatorDataFeed(bt.feeds.PandasData):
    lines = ('rsi', 'macd', 'bbl', 'bbu', 'atr', 'vix')
    params = (
        ('datetime', None),
        ('open', 'SPY_Open'),
        ('high', 'SPY_High'),
        ('low', 'SPY_Low'),
        ('close', 'SPY_Close'),
        ('volume', 'SPY_Volume'),
        ('rsi', 'SPY_RSI_14'),
        ('macd', 'SPY_MACD_12_26'),
        ('bbl', 'SPY_BBL_20_2'),
        ('bbu', 'SPY_BBU_20_2'),
        ('atr', 'SPY_ATR_14'),
        ('vix', 'VIXCLS')
    )
```

```python
# --- CELL 5: TRADING STRATEGY ---
class MultiIndicatorStrategy(bt.Strategy):
    params = (
        ('rsi_oversold', 35),
        ('rsi_overbought', 65),
        ('vix_entry', 25),
        ('vix_exit', 30),
        ('atr_multiplier', 1.5),
        ('bb_period', 20))
    
    def __init__(self):
        # Indicator tracking
        self.ind = {}
        for data in self.datas:
            self.ind[data._name] = {
                'rsi': data.rsi,
                'macd': data.macd,
                'bbl': data.bbl,
                'bbu': data.bbu,
                'atr': data.atr,
                'vix': data.vix
            }
        
        # Trade journal
        self.trade_log = []
        
    def next(self):
        for data in self.datas:
            ticker = data._name
            close = data.close[0]
            ind = self.ind[ticker]
            
            # ---- Entry Conditions ----
            entry_cond = (
                ind['rsi'][0] < self.p.rsi_oversold and
                ind['macd'][0] > ind['macd'][-1] and  # MACD rising
                ind['vix'][0] < self.p.vix_entry and
                close < ind['bbl'][0]  # Price below lower Bollinger Band
            )
            
            # ---- Exit Conditions ----
            exit_cond = (
                ind['rsi'][0] > self.p.rsi_overbought or
                ind['macd'][0] < ind['macd'][-1] or
                ind['vix'][0] > self.p.vix_exit or
                close > ind['bbu'][0]  # Price above upper Bollinger Band
            )
            
            # ---- Trade Execution ----
            position = self.getposition(data)
            if not position:
                if entry_cond:
                    size = self.calculate_position_size(data)
                    self.buy(data=data, size=size)
                    self.log_trade('ENTER', ticker, close)
            else:
                if exit_cond:
                    self.close(data=data)
                    self.log_trade('EXIT', ticker, close)

    def calculate_position_size(self, data):
        risk_per_trade = self.broker.getvalue() * 0.02  # 2% risk
        atr = self.ind[data._name]['atr'][0]
        return int(risk_per_trade / (atr * self.p.atr_multiplier))

    def log_trade(self, action, ticker, price):
        self.trade_log.append({
            'date': self.datetime.date(),
            'action': action,
            'ticker': ticker,
            'price': price,
            'portfolio_value': self.broker.getvalue()
        })
```

```python
# --- CELL 6: VISUALIZATION ---
def plot_indicators(data, ticker='SPY', days=365):
    fig, axs = plt.subplots(5, 1, figsize=(16, 20))
    
    # Price & Bollinger Bands
    axs[0].plot(data[f'{ticker}_Close'], label='Price')
    axs[0].plot(data[f'{ticker}_BBL_20_2'], label='Lower Bollinger')
    axs[0].plot(data[f'{ticker}_BBU_20_2'], label='Upper Bollinger')
    axs[0].set_title(f'{ticker} Price & Bollinger Bands')
    axs[0].legend()
    
    # RSI
    axs[1].plot(data[f'{ticker}_RSI_14'], label='RSI', color='purple')
    axs[1].axhline(30, linestyle='--', color='green')
    axs[1].axhline(70, linestyle='--', color='red')
    axs[1].set_title('RSI (14)')
    
    # MACD
    axs[2].plot(data[f'{ticker}_MACD_12_26'], label='MACD', color='orange')
    axs[2].set_title('MACD (12,26,9)')
    
    # ATR
    axs[3].plot(data[f'{ticker}_ATR_14'], label='ATR', color='brown')
    axs[3].set_title('Average True Range (14)')
    
    # VIX
    axs[4].plot(data['VIXCLS'], label='VIX', color='black')
    axs[4].set_title('CBOE Volatility Index (VIX)')
    
    plt.tight_layout()
    plt.savefig('technical_indicators.png')
    plt.show()

# Generate visualization
plot_indicators(full_data[-365:])  # Last year of data
```

```python
# --- CELL 7: BACKTEST SETUP ---
def configure_cerebro():
    cerebro = bt.Cerebro(stdstats=False)
    
    # Add data feeds
    for ticker in ['SPY', 'QQQ']:
        data = IndicatorDataFeed(
            dataname=full_data,
            name=ticker,
            open=f'{ticker}_Open',
            high=f'{ticker}_High',
            low=f'{ticker}_Low',
            close=f'{ticker}_Close',
            volume=f'{ticker}_Volume',
            rsi=f'{ticker}_RSI_14',
            macd=f'{ticker}_MACD_12_26',
            bbl=f'{ticker}_BBL_20_2',
            bbu=f'{ticker}_BBU_20_2',
            atr=f'{ticker}_ATR_14',
            vix='VIXCLS'
        )
        cerebro.adddata(data)
    
    # Configure strategy and broker
    cerebro.addstrategy(MultiIndicatorStrategy)
    cerebro.broker.setcash(100000)
    cerebro.broker.setcommission(commission=0.0005)  # 0.05% commission
    cerebro.addsizer(bt.sizers.PercentSizer, percents=98)
    
    # Add analyzers
    cerebro.addanalyzer(bt.analyzers.PyFolio, _name='pyfolio')
    cerebro.addanalyzer(bt.analyzers.TradeAnalyzer, _name='trades')
    
    return cerebro
```

```python
# --- CELL 8: RUN BACKTEST ---
cerebro = configure_cerebro()
print(f'Initial Portfolio Value: {cerebro.broker.getvalue():,.2f}')
results = cerebro.run()
print(f'Final Portfolio Value: {cerebro.broker.getvalue():,.2f}')
```

```python
# --- CELL 9: TRADE ANALYSIS ---
def analyze_results(results):
    strat = results[0]
    
    # Trade Log
    trades = pd.DataFrame(strat.trade_log)
    if not trades.empty:
        print("\nTrade History:")
        display(trades)
        
        # Plot Portfolio Growth
        plt.figure(figsize=(12,6))
        plt.plot(trades['date'], trades['portfolio_value'])
        plt.title('Portfolio Value Over Time')
        plt.xlabel('Date')
        plt.ylabel('Value ($)')
        plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m'))
        plt.gcf().autofmt_xdate()
        plt.savefig('portfolio_growth.png')
    else:
        print("No trades executed - check indicator signals")
    
    # Strategy Metrics
    try:
        trade_analysis = strat.analyzers.trades.get_analysis()
        print(f"\nTotal Trades: {trade_analysis['total']['total']}")
        print(f"Win Rate: {trade_analysis['won']['total']/trade_analysis['total']['total']:.2%}")
    except:
        print("\nCould not calculate performance metrics")

analyze_results(results)
```

```python
# --- CELL 10: SENSITIVITY ANALYSIS ---
def optimize_parameters():
    cerebro.optstrategy(
        MultiIndicatorStrategy,
        rsi_oversold=range(25, 36, 5),
        vix_entry=range(15, 26, 5),
        atr_multiplier=[1.0, 1.5, 2.0]
    )
    
    opt_results = cerebro.run()
    
    # Find best parameter set
    best_sharpe = -np.inf
    for result in opt_results:
        sharpe = result.analyzers.sharperatio.get_analysis()['sharperatio']
        if sharpe > best_sharpe:
            best_params = result.params
            best_sharpe = sharpe
    
    print(f"\nBest Parameters: {best_params}")
    print(f"Best Sharpe Ratio: {best_sharpe:.2f}")

# Uncomment to run optimization
# optimize_parameters()
```
